# Análise exploratória dos dados e cruzamento de informações

Neste notebook serś feito a análise exploratóri dos dados para tirar insigth do dataset em que foram selecionadas as colunas\

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
url = 'https://raw.githubusercontent.com/gustavoramos82/maratona-tech-rs/main/Datasets%20Utilizados/barragem_rs_colunas_selecionadas.csv'
barragem_rs = pd.read_csv(url,encoding='latin1')
barragem_rs.drop('Unnamed: 0',axis=1,inplace=True)
barragem_rs.head()
```

```python
barragem_rs.drop_duplicates()
```

```python
empree_ger = pd.read_csv('/content/gera_empre.csv')
empree_ger.drop('_id',axis=1,inplace=True)
empree_ger.head()
```

```python
empree_ger = empree_ger[empree_ger['SigUFPrincipal']=='RS']
empree_ger.head()
```

```python
cruz_barra_rs = barragem_rs.merge(empree_ger,on='IdeNucleoCEG')
cruz_barra_rs.head()
```

```python
cruz_barra_rs['NumCoordNEmpreendimento'] = cruz_barra_rs['NumCoordNEmpreendimento'].str.replace(',','.')
cruz_barra_rs['NumCoordNEmpreendimento'] = cruz_barra_rs['NumCoordNEmpreendimento'].astype(float)
cruz_barra_rs['NumCoordEEmpreendimento'] = cruz_barra_rs['NumCoordEEmpreendimento'].str.replace(',','.')
cruz_barra_rs['NumCoordEEmpreendimento'] = cruz_barra_rs['NumCoordEEmpreendimento'].astype(float)
```

```python
fig = px.scatter_mapbox(cruz_barra_rs, lat="NumCoordNEmpreendimento", lon="NumCoordEEmpreendimento",
                        hover_name="NomUsina", hover_data=["NomEmpresa", "NomBarramento"],
                       zoom=3, height=300,color='DscDanoPotencialGeral')
fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```

```python
cruz_barra_rs.info()
```

```python
cruz_barra_rs.drop(['DatGeracaoConjuntoDados','NomEmpreendimento','CodCEG',
                    'DatEntradaOperacao','DatInicioVigencia','DatFimVigencia'],axis=1,
                   inplace=True)
cruz_barra_rs.head()
```

```python
cruz_barra_rs['NomFonteCombustivel'].unique()
cruz_barra_rs['DscPropriRegimePariticipacao'].unique()
cruz_barra_rs.drop(['NomFonteCombustivel',
                    'DscPropriRegimePariticipacao'],axis=1,inplace=True)
cruz_barra_rs.head()
```

```python
fig = px.scatter_mapbox(cruz_barra_rs, lat="NumCoordNEmpreendimento", lon="NumCoordEEmpreendimento",
                        hover_name="NomUsina", hover_data=["NomEmpresa", "NomBarramento"],
                       zoom=3, height=300,color='DscCategoriaRiscoGeral')
fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```

```python
cruz_barra_rs.to_csv('fiscali-barragem-dado-ele.csv')
```
