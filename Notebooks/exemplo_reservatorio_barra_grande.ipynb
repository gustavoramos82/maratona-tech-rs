```python
import pandas as pd
```

```python
defluencia = pd.read_csv('/content/drive/MyDrive/Datasets/Reservatorio Federal.csv')
defluencia.drop('Unnamed: 0',axis=1,inplace=True)
defluencia.head()
```

```python
defluencia = defluencia[['data','id_reservatorio','nome_reservatorio','defluencia']]
defluencia.head()
```

```python
defluencia.columns = ['data','codigo','no','defluencia']
defluencia.head()
```

```python
lista_reservatorio = pd.read_csv('/content/lista_reservatorios.csv')
lista_reservatorio.drop(['Unnamed: 0','codigo_municipio_ibge','estado_nome','municipio'],
                        axis=1,inplace=True)
lista_reservatorio.head()
```

```python
defluencia = defluencia.merge(lista_reservatorio,on='codigo')
defluencia = defluencia[defluencia['estado_sigla']=='RS']
defluencia.head()
```

```python
defluencia.drop(['no','estado_sigla'],axis=1,inplace=True)
defluencia.head()
```

```python
defluencia.reset_index(inplace=True)
defluencia.drop('index',axis=1,inplace=True)
defluencia.head()
```

```python
defluencia['reservatorio'].unique()
```

```python
defluencia = defluencia[defluencia['reservatorio']=='BARRA GRANDE']
defluencia.head()
```

```python
defluencia.iloc[:,0:3].to_csv('Reser barra grande.csv')
```
