```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'https://raw.githubusercontent.com/gustavoramos82/maratona-tech-rs/main/Datasets%20Utilizados/fiscalizacao-seguranca-barragem.csv'
barragem = pd.read_csv(url,encoding='latin1',
                       sep=';')
barragem.head()
```

```python
barragem.shape
```

```python
barragem.isna().sum()[0:31]
```

```python
barragem.isna().sum()[61:87]
```

Podemos ver que temos colunas com muito campos em branco, então as mesmas serão removidas

```python
barragem.drop(['DatInicioConstrucao','MdaLatitude','MdaLongitude','DatPrevistaConclusaoPSB','DatPrevistaConclusaoPAE',
               'DatPrevistaConclusaoISR','DatPrevistaConclusaoISE'],
              axis=1,inplace=True)
barragem.head()
```

```python
# filtrar para o rio grande do sul
barragem_rs = barragem[barragem['SigEstadoUsina']=='RS']
barragem_rs.head()
```

```python
barragem_rs.reset_index(inplace=True)
barragem_rs.drop('index',axis=1,inplace=True)
barragem_rs.head()
```

```python
barragem_rs.info()
```

```python
barragem_rs.drop(['DatConclusaoISE','DscFinalidade'],axis=1,inplace=True)
barragem_rs.head()
```

```python
barragem_rs.tail()
```

```python
barragem_rs['DatGeracaoConjuntoDados'].unique()
```

Removeremos a coluna 'DatGeracaoConjuntoDados' pois está marca na data 01-05-2024 logo não será interessante ter essa coluna

```python
barragem_rs.drop('DatGeracaoConjuntoDados',axis=1,inplace=True)
barragem_rs.head()
```
